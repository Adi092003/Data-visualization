```python
import pandas as pd

# Load the dataset
matches = pd.read_csv("matches.csv")
```

```python
import pandas as pd

# Calculate the number of tosses won by each team
toss_count = matches['toss_winner'].value_counts()

# Create a new DataFrame
new_dataset = pd.DataFrame({
    'Team': toss_count.index,
    'Tosses_Won': toss_count.values
})

# Transpose the DataFrame
new_dataset = new_dataset.T

# Print the new dataset
print(new_dataset)
```

```python
import matplotlib.pyplot as plt

# Assuming 'new_dataset' is the DataFrame containing team names and toss counts
# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(new_dataset.loc['Tosses_Won'], labels=new_dataset.loc['Team'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Tosses Won by Each Team in IPL')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'new_dataset' is the DataFrame containing team names and toss counts

# Create a bar chart
plt.figure(figsize=(10, 8))
plt.bar(new_dataset.loc['Team'], new_dataset.loc['Tosses_Won'], color='skyblue')
plt.xlabel('Team')
plt.ylabel('Tosses Won')
plt.title('Tosses Won by Each Team in IPL')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'new_dataset' is the DataFrame containing team names and toss counts

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(new_dataset.loc['Team'], new_dataset.loc['Tosses_Won'], color='lightgreen')
plt.xlabel('Tosses Won')
plt.ylabel('Team')
plt.title('Tosses Won by Each Team in IPL')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
